```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/wuzhong-zhu/fp/main/foodpanda_data.csv"

# Assign colum names to the dataset
names = ['cuisine', 'orderCnt', 'customerID']

# Read dataset to pandas dataframe
dataset = pd.read_csv(url)
```

```python
dataset
```

```python
cols = dataset["cuisine"].unique()
IDs = dataset["Customer_id"].unique()

cols = np.insert(cols,0,["customerID"])
df = pd.DataFrame(columns=cols)

for i in IDs:
    df = df.append({"customerID": i}, ignore_index=True)

for index, row in dataset.iterrows():
    df.loc[df.customerID == row['Customer_id'],row['cuisine']] = row['Numbers of order']

df
```

```python
cols = dataset["cuisine"].unique()
cols = np.insert(cols, 0, ["result"])
cols = np.insert(cols, 0, ["customerID"])
df2 = pd.DataFrame(columns=cols)

for index, row in dataset.iterrows():
    for i in range(row['Numbers of order']):
        df2 = df2.append({
            "result":row['cuisine'],
            "customerID":row["Customer_id"]}, ignore_index=True)

df2
```

```python
df2.to_csv('save2.csv')
df.to_csv('save1.csv')
```

```python

for index, row in df2.iterrows():
#     print(row['customerID'])
        temp =  df.loc[df.customerID == row['customerID']]
        temp2 = temp.to_numpy()[0][1:]
        row[2:] = temp2
df2
```

```python
df2 = df2.fillna(0)
df = df.fillna(0)
df2.to_csv('save2.csv')
df.to_csv('save1.csv')
```

```python
X = df2.iloc[:, 2:].values
y = df2.iloc[:, 1].values
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
KNN below:
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
temp = df.drop('customerID', 1)
y_pred = classifier.predict(temp)
print(y_pred)
np.savetxt("KNN result.csv", y_pred, fmt='%s')
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
Decision tree below
```

```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
temp = df.drop('customerID', 1)
y_pred = clf.predict(temp)
print(y_pred)
np.savetxt("decision tree result.csv", y_pred, fmt='%s')
```
